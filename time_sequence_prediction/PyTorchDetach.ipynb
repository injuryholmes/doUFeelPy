```python
import torch as T
from torchviz import *
```

# Fully attached graph

```python
x=T.ones(10, requires_grad=True)

y=x**2
z=x**3
r=(y+z).sum()
make_dot(r)
```

```python
r.requires_grad
```

```python
y.requires_grad
```

```python
z.requires_grad
```

```python
r.backward()
x.grad
```

# Partially Detached Graph

```python
x=T.ones(10, requires_grad=True)
y=x**2
z=x.detach()**3
r=(y+z).sum()    
make_dot(r)
```

```python
r.backward()
x.grad
```

```python
z.requires_grad
```
